### Contour detection 

```python
img = cv2.pyrDown(cv2.imread('hammer.PNG', cv2.IMREAD_UNCHANGED))
plt.imshow(img)
plt.axis('off');
```

```python
ret, thresh = cv2.threshold(cv2.cvtColor(img.copy(), cv2.COLOR_BGR2GRAY), 127, 255, cv2.THRESH_BINARY)
image, contours, hier = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
for c in contours:
    # find bounding box coordinates
    x, y, w, h = cv2.boundingRect(c)
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 1)
    # find minimum area
    rect = cv2.minAreaRect(c)
    # calculate coordinates of the minimum area rectangle
    box = cv2.boxPoints(rect)
    # normalize coordinates to integers
    box = np.int0(box)
    # draw contours
    cv2.drawContours(img, [box], 0, (0, 0, 255), 1)
    
    # calculate center and radius of minimum enclosing circle
    (x, y), radius = cv2.minEnclosingCircle(c)
    # cast to integers
    center = (int(x), int(y))
    radius = int(radius)
    # draw the circle
    img = cv2.circle(img, center, radius, (0, 255, 0), 1)
   
plt.imshow(img)
plt.axis('off');
```

### Example of foreground detection with GrabCut

```python
img = cv2.imread('angel.PNG', cv2.IMREAD_UNCHANGED)
plt.imshow(img)
plt.axis('off');
```

```python
img.shape
```

```python
img = img[:, :, :3]
```

```python
mask = np.zeros(img.shape[:2], np.uint8)

bgdModel = np.zeros((1, 65), np.float64)
fgdModel = np.zeros((1, 65), np.float64)

rect = (100, 50, 421, 378)
cv2.grabCut(img, mask, rect, bgdModel, fgdModel, 5, cv2.GC_INIT_WITH_RECT)

mask2 = np.where((mask == 2) | (mask == 0), 0, 1).astype('uint8')
img = img * mask2[:, :, np.newaxis]

plt.subplot(121), plt.imshow(img)
plt.title("grabcut"), plt.xticks([]), plt.yticks([])
plt.subplot(122),
plt.imshow(cv2.cvtColor(cv2.imread('angel.PNG'), cv2.COLOR_BGR2RGB))
plt.title("original");
plt.xticks([]);
plt.yticks([]);
```

### Image segmentation with the Watershed algorithm

```python
img = cv2.imread('basil.PNG', cv2.IMREAD_UNCHANGED)
img = img[:, :, :3]
plt.imshow(img)
plt.axis('off');
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.gray()
plt.imshow(gray)
plt.axis('off');
```

```python
# this operation helps dividing the image in two, blacks and whites
ret, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV+ cv2.THRESH_OTSU)
# remove noise
kernel = np.ones((3, 3), np.uint8)
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations = 2)
# dilating and then eroding an image to extract features
sure_bg = cv2.dilate(opening, kernel, iterations = 3)
# obtain foreground
dist_transform = cv2.distanceTransform(opening, cv2.DIST_L2, 5)
ret, sure_fg = cv2.threshold(dist_transform, 0.7 * dist_transform.max(), 255, 0)
sure_fg = np.uint8(sure_fg)
unknown = cv2.subtract(sure_bg, sure_fg)
# build "barriers" to stop the water from merging
ret, markers = cv2.connectedComponents(sure_fg)
markers = markers + 1
markers[unknown == 255] = 0
# let the water fall and our barriers be drawn in red
markers = cv2.watershed(img, markers)
img[markers == -1] = [255, 0, 0]
plt.imshow(img)
plt.axis('off');
```

### Feature detection algorithms
#### Detecting features - corners

```python
img = cv2.imread('chess_board.PNG')
plt.imshow(img)
plt.axis('off');
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)
dst = cv2.cornerHarris(gray, 2, 23, 0.04)
img[dst > 0.01 * dst.max()] = [255, 0, 0]
plt.imshow(img)
plt.axis('off');
```

```python
img = cv2.imread('chess_board.PNG')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)
dst = cv2.cornerHarris(gray, 2, 3, 0.04)
img[dst > 0.01 * dst.max()] = [255, 0, 0]
plt.imshow(img)
plt.axis('off');
```

#### Feature matching with ORB --- match the logo of a band

```python
import cv2
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('manowar_logo.PNG', cv2.IMREAD_GRAYSCALE)
plt.imshow(img1)
plt.axis('off');
```

```python
img2 = cv2.imread('manowar_single.PNG', cv2.IMREAD_GRAYSCALE)
plt.imshow(img2)
plt.axis('off');
```

```python
orb = cv2.ORB_create()
kp1, des1 = orb.detectAndCompute(img1, None)
kp2, des2 = orb.detectAndCompute(img2, None)

bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck = True)

matches = bf.match(des1, des2)
matches = sorted(matches, key = lambda x: x.distance)
img3 = cv2.drawMatches(img1, kp1, img2, kp2, matches[:40], img2, flags = 2)
plt.imshow(img3)
plt.axis('off');
```

#### People detection

```python
def is_inside(o, i):
    ox, oy, ow, oh = o
    ix, iy, iw, ih = i
    return ox > ix and oy > iy and ox + ow < ix + iw and oy + oh < iy + ih

def draw_person(image, person):
    x, y, w, h = person
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 255), 2)
```

```python
img = cv2.imread('person.PNG', 0)
plt.gray()
plt.imshow(img)
plt.axis('off');
```

```python
# load model to detect face 
face_cascade = cv2.CascadeClassifier('./face_detection/cascades/haarcascade_frontalface_default.xml')
faces = face_cascade.detectMultiScale(img, 1.3, 5)
# draw rectangle
for (x, y, w, h) in faces:
    img = cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)  
        
plt.imshow(img)
plt.axis('off');
```
